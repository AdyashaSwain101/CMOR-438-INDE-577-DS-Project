## Logistic Regression

#### Logistic Regression: A supervised learning algorithm used for binary classification that models the probability of a class using the logistic (sigmoid) function.
#### ![logistic regression](https://miro.medium.com/v2/resize:fit:460/1*klFuUpBGVAjTfpTak2HhUA.png)

#### Importing libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import numpy as np
```

```python

data = pd.read_csv('cleaned_housing.csv')
```

```python

X = data.drop('MEDV', axis=1)
y = data['MEDV']
```

```python

median_value = y.median()
y_binary = (y > median_value).astype(int)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)
```

```python

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python

logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train_scaled, y_train)

```

```python

y_pred = logistic_model.predict(X_test_scaled)

```

```python

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(report)
```

```python

feature_of_interest = 0  
x_range = np.linspace(X_train_scaled[:, feature_of_interest].min(), X_train_scaled[:, feature_of_interest].max(), 100)


X_dummy = np.tile(X_train_scaled.mean(axis=0), (100, 1))
X_dummy[:, feature_of_interest] = x_range


y_proba = logistic_model.predict_proba(X_dummy)[:, 1]

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_proba)
plt.xlabel(f'Feature {X.columns[feature_of_interest]}')
plt.ylabel('Probability')
plt.title('Probability vs. Feature Plot')
plt.show()

```
