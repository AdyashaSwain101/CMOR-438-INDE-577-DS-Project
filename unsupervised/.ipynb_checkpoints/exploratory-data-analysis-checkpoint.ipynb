```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pywaffle import Waffle
import os
```

```python
path = os.getcwd() + '/data/'
```

```python
df = pd.read_csv(path+'preprocessed.csv')
```

```python
variable_counts = df['Variable'].value_counts()

# Set Seaborn style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=variable_counts.index, y=variable_counts.values, palette="Set1")

# Adding titles and labels with Seaborn's themes
plt.title('Distribution of Fuel Types', fontsize=16)
plt.xlabel('Variable', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter for Electricity generation and Fuel
df_filtered = df[(df['Category'] == 'Electricity generation') & (df['Subcategory'] == 'Fuel')]

# Group by Variable and sum the Values
grouped_data = df_filtered.groupby('Variable')['Value'].sum().sort_values(ascending=False)

# Prepare the data for the waffle chart
waffle_data = grouped_data.to_dict()

# Calculate the total value
total_value = sum(waffle_data.values())

# Normalize the values to percentages
waffle_data = {k: v / total_value * 100 for k, v in waffle_data.items()}

# Generate a color palette with the correct number of colors
colors = sns.color_palette("husl", n_colors=len(waffle_data))

# Create the waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    values=waffle_data,
    title={'label': 'Distribution of Fuel Types in Electricity Generation', 'fontsize': 16},
    labels=[f"{k} ({v:.1f}%)" for k, v in waffle_data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.2), 'ncol': 2, 'fontsize': 8},
    colors=colors,
    figsize=(12, 10)
)

# Add a text annotation for the total
plt.text(0.95, 0.05, f'Total Electricity Generation: {total_value:.2f} TWh', 
         horizontalalignment='right', 
         verticalalignment='bottom', 
         transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for electricity imports
df_electricity_imports = df[df['Category'] == 'Electricity imports']

# Aggregate total electricity imports by country
country_imports = df_electricity_imports.groupby('Area')['Value'].sum().sort_values(ascending=False)[:20]

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# Create a horizontal bar plot for countries with the highest electricity imports
plt.figure(figsize=(12, 10))
ax = sns.barplot(x=country_imports.values, y=country_imports.index, palette="coolwarm_r")

# Add value labels
for i, v in enumerate(country_imports.values):
    ax.text(v, i, f'{v:.2f}', va='center', fontweight='bold')

# Add titles and labels
plt.title('Top 20 Countries by Electricity Imports', fontsize=18, fontweight='bold')
plt.xlabel('Total Electricity Imports (TWh)', fontsize=14)
plt.ylabel('Country', fontsize=14)

# Remove top and right spines
sns.despine(top=True, right=True)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for electricity demand
# Assuming df is your DataFrame
# Filter the DataFrame for electricity demand
df_demand = df[df['Category'] == 'Electricity demand']

# Aggregate total demand and demand per capita by country
total_demand = df_demand[df_demand['Variable'] == 'Demand'].groupby('Area')['Value'].sum()
per_capita_demand = df_demand[df_demand['Variable'] == 'Demand per capita'].groupby('Area')['Value'].mean()

# Combine the data
combined_demand = pd.DataFrame({'Total Demand': total_demand, 'Per Capita Demand': per_capita_demand})
combined_demand = combined_demand.dropna().sort_values('Total Demand', ascending=False)

# Select top 30 countries by total demand
top_30 = combined_demand.head(30)

# Create the scatter plot
fig, ax = plt.subplots(figsize=(16, 10))
scatter = sns.scatterplot(data=top_30, x='Total Demand', y='Per Capita Demand', 
                size='Total Demand', sizes=(20, 2000), 
                alpha=0.6, palette='viridis', ax=ax)

# Annotate points with country names
for idx, row in top_30.iterrows():
    ax.text(row['Total Demand'], row['Per Capita Demand'], idx, 
            fontsize=9, ha='right', va='center', color='black',
            bbox=dict(facecolor='white', alpha=0.5, boxstyle='round,pad=0.3'))

# Set labels and title
ax.set_xlabel('Total Electricity Demand (TWh)', fontsize=12)
ax.set_ylabel('Electricity Demand per Capita (MWh)', fontsize=12)
ax.set_title('Electricity Demand: Total vs Per Capita (Top 30 Countries)', fontsize=16)

# Use log scale for x-axis due to large range of values
ax.set_xscale('log')

# Add a colorbar legend
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=top_30['Total Demand'].min(), 
                                                              vmax=top_30['Total Demand'].max()))
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax)
cbar.set_label('Total Electricity Demand (TWh)', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for electricity generation and emissions
df_generation = df[(df['Category'] == 'Electricity generation') & (df['Variable'] == 'Total Generation')]
df_emissions = df[(df['Category'] == 'Power sector emissions') & (df['Variable'] == 'Total emissions')]
df_intensity = df[(df['Category'] == 'Power sector emissions') & (df['Variable'] == 'CO2 intensity')]

# Aggregate data by country
generation = df_generation.groupby('Area')['Value'].sum()
emissions = df_emissions.groupby('Area')['Value'].sum()
intensity = df_intensity.groupby('Area')['Value'].mean()

# Combine the data
combined_data = pd.DataFrame({
    'Generation': generation,
    'Emissions': emissions,
    'CO2 Intensity': intensity
}).dropna()

# Sort by emissions and get top 30 countries
top_30 = combined_data.sort_values('Emissions', ascending=False).head(30)

# Create the scatter plot
plt.figure(figsize=(16, 10))
scatter = plt.scatter(top_30['Generation'], top_30['Emissions'], 
                      s=top_30['Generation']/50, # Size based on generation
                      c=top_30['CO2 Intensity'], # Color based on CO2 intensity
                      cmap='YlOrRd', alpha=0.7)

# Annotate points with country names
for idx, row in top_30.iterrows():
    plt.annotate(idx, (row['Generation'], row['Emissions']), 
                 xytext=(5,5), textcoords='offset points', fontsize=8, 
                 bbox=dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9))

# Set labels and title
plt.xlabel('Electricity Generation (TWh)', fontsize=12)
plt.ylabel('CO2 Emissions (Mt)', fontsize=12)
plt.title('Electricity Generation vs CO2 Emissions (Top 30 Emitting Countries)', fontsize=16)

# Add a colorbar legend for CO2 intensity
cbar = plt.colorbar(scatter)
cbar.set_label('CO2 Intensity (g/kWh)', fontsize=10)

# Add a legend for bubble size
sizes = [1000, 5000, 10000]
labels = ['1000 TWh', '5000 TWh', '10000 TWh']
legend_elements = [plt.scatter([], [], s=size/50, c='gray', alpha=0.7, label=label)
                   for size, label in zip(sizes, labels)]
plt.legend(handles=legend_elements, title='Electricity Generation', 
           loc='upper left', title_fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for electricity generation from green sources
green_sources = ['Hydro', 'Wind', 'Solar', 'Geothermal', 'Bioenergy']
df_green = df[(df['Category'] == 'Electricity generation') & (df['Variable'].isin(green_sources))]

# Aggregate total green energy generation by country
green_energy = df_green.groupby('Area')['Value'].sum().sort_values(ascending=False)

# Select top 30 countries by green energy generation
top_30_green = green_energy.head(30)

# Create a bar plot for countries with the highest green energy generation
plt.figure(figsize=(14, 10))
sns.barplot(x=top_30_green.values, y=top_30_green.index, palette="summer")

# Add titles and labels
plt.title('Top 30 Countries by Green Energy Generation', fontsize=16)
plt.xlabel('Total Green Energy Generation (TWh)', fontsize=14)
plt.ylabel('Country', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for power sector emissions
df_emissions = df[(df['Category'] == 'Power sector emissions') & (df['Variable'] == 'Total emissions')]

# Aggregate emissions by country
emissions = df_emissions.groupby('Area')['Value'].sum().sort_values()

# Select bottom 20 countries (excluding countries with zero emissions)
bottom_20 = emissions[emissions > 0].head(20)

# Create the bar plot
plt.figure(figsize=(12, 10))
ax = sns.barplot(x=bottom_20.values, y=bottom_20.index, palette="YlOrRd_r")

# Add value labels to the end of each bar
for i, v in enumerate(bottom_20.values):
    ax.text(v, i, f' {v:.2f}', va='center')

# Set labels and title
plt.xlabel('Total CO2 Emissions from Power Sector (Mt)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('20 Countries with Lowest Power Sector CO2 Emissions', fontsize=16)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
